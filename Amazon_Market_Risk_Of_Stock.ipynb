* To calculate the market risk of a stock, you can calculate its beta value. Beta measures the volatility of a stock relative to the overall market. A beta value of 1 indicates that the stock's volatility is equivalent to the market's volatility, while a beta value greater than 1 indicates that the stock is more volatile than the market, and a beta value less than 1 indicates that the stock is less volatile than the market.

```python
import yfinance as yf
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ["AMZN"]
```

```python
data = yf.download(tickers,start = "2015-01-01")
data
```

```python
start_date = '2015-01-01'
end_date  = pd.Timestamp.today().strftime('%Y-%M-%d')
```

```python
# Calculate the daily returns
daily_returns = data['Adj Close'].pct_change()
```

```python
# Get the S&P 500 index data
tickers = ["GOOGL"]
data_google = yf.download(tickers,start = "2015-01-01")
data_google
```

```python
# Calculate the daily returns of the index
index_returns = data_google['Adj Close'].pct_change()
```

```python
#Merge the stock and index returns into a single DataFrame
returns = pd.concat([daily_returns, index_returns], axis=1, join='inner')
```

```python
# Calculate the covariance matrix and beta value
covariance_matrix = returns.cov()
beta = covariance_matrix.iloc[0, 1] / covariance_matrix.iloc[1, 1]
```

```python
# Print the beta value
print(f'Beta of Amazon: {beta:.2f}')
```

* A beta of 0.77 means that the stock of Amazon is 23% less volatile than the overall market.
